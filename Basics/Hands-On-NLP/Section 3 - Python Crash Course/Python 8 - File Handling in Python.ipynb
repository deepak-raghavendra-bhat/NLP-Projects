File I/O

Writing to a file

```python
inp = input('Enter some text: ')
```

```python
with open('textFile.txt','a') as f:
    f.write(inp)
    
# Reading from a file
    
with open('textFile.txt','r') as f:
    string = f.read()
    print(string)
```
